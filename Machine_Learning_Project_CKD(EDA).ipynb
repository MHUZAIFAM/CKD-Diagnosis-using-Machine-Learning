```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading and loading the dataset
df = pd.read_csv('kidney_disease_preprocessed.csv')
print("Dataset Loaded Success")
```

```python
#Printing the first five rows of the dataset
df.head()
```

```python
#Getting the summary of the dataset
df.info()
```

```python
#Now check for missing values
df.isnull().sum()
```

**Visualizing the Dataset**

Histograms

```python
#Histogram for the Dataset
histograms = df.hist(figsize=(10,10),bins = 20, color = 'red', edgecolor = 'black')
plt.tight_layout()  # Adjust layout to avoid overlap
plt.show()
```

Box PLots

```python
#Drawing Box Plots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['age', 'bp', 'sg','al']])
plt.title('Box Plot for Features')
plt.xticks(rotation=90)
plt.show()
```

**Class Distribution (Bar/Pie Chart)**

Bar Plot

```python
# Visualize class distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='classification', data=df, palette='pastel')
plt.title('Class Distribution')
plt.show()
```

PIE Chart

```python
#To make a pie chart
target_counts = df['classification'].value_counts()
plt.figure(figsize=(8, 8))
plt.title('Class Distribution')
plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

**Correlation Matrix (Heatmap)**

```python
# Correlation Matrix (Heatmap)
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

**Scatter Plot**

```python
#To create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='bp', hue='classification', data=df)
plt.title('Scatter Plot of Age vs. Blood Pressure')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='sg', hue='classification', data=df)
plt.title('Scatter Plot of Age vs. SG')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='al', hue='classification', data=df)
plt.title('Scatter Plot of Age vs. AL')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bp', y='rbc', hue='classification', data=df)
plt.title('Scatter Plot of Blood Pressure vs. SG')
```
